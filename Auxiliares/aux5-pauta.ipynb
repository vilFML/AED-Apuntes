##### a)
Aidan necesita que programes una función def getKthElement(arr, k) que dado un arreglo, entregue el k-ésimo elemento.

> En un ABB, hacer un recorrido **in-orden** accede ascendentemente a las llaves almacenadas.

- La idea es usar el recorrido in-orden de un ABB para acceder a las llaves ascendentemente.
- Además, se puede llevar la cuenta del $k$ésimo elemento con una variable $k$, que comienza en 1 y aumentará de valor hasta llegar al valor deseado.



Primero, consideramos el ABB y su Nodo.

```python
!pip install aed-utilities
import aed_utilities as aed

class Nodo:
  def __init__(self, info, izq = None, der = None):
    self.info = info
    self.izq = izq
    self.der = der
    self.num = 1 # tamaño del arbol

class ABB:
  def __init__(self):
    self.raiz = None
  def insertar(self, info):
    def _insertar(nodo, info):
      if nodo == None:
        return Nodo(info)
      if nodo.info < info:
        nodo.der = _insertar(nodo.der, info)
      else:
        nodo.izq = _insertar(nodo.izq, info)
      return nodo
    self.raiz = _insertar(self.raiz, info)
  def dibujar(self):
    BTdraw = aed.BinaryTreeDrawer(fieldData="info", fieldLeft="izq", fieldRight = "der", drawNull=True)
    BTdraw.draw_tree(self, "raiz")
```

```python
# Ahora, el arreglo lo pasamos al ABB para que "se ordene"
import numpy as np

arr =  np.random.choice(range(1,200), 10, replace=False)
print(arr)
arbol = ABB()
for el in arr:
  arbol.insertar(el)
arbol.dibujar()
```

La solución:
El Contador $k$ como "Vidas" o "Pasos". Como queremos el $k$-ésimo, no necesitamos guardar todos los elementos en una lista nueva.
Usamos la variable $k$ como un contador regresivo:
- Paso A: Bajamos por la izquierda hasta el más pequeño.
- Paso B: Cada vez que "procesamos" un nodo (cuando ya regresamos de su hijo izquierdo), decimos: "Ya encontré uno más", y restamos 1 a $k$.
- Paso C: En el momento exacto en que $k$ llega a 0, el nodo en el que estamos parados es nuestra respuesta.

```python
def getKth(self, k):
  def _getKth(nodo, k):
    if nodo == None:
      return k
    else:
      #Recorrido inorden
      k = _getKth(nodo.izq, k) #primero izq
      if k == 0:
        return k
      k -= 1 #contar el elemento central k = k - 1
      if k == 0:
        self.k = nodo.info
        return k
      return _getKth(nodo.der, k) #despues el arbol derecho
  #end _getKth
  if _getKth(self.raiz, k) == 0:
    return self.k
  return -1

ABB.getKth = getKth

for i in range(1,12):
  print(arbol.getKth(i))
#Cuanto tarda este algoritmo?
#O(k) O(n) k=n => O(n)
```

```python
# Variación usando numero de nodos
def insertar2(self, info):
  def _insertar(nodo, info):
    if nodo == None:
      return Nodo(info)
    nodo.num += 1
    if nodo.info < info:
      nodo.der = _insertar(nodo.der, info)
    else:
      nodo.izq = _insertar(nodo.izq, info)
    return nodo
  self.raiz = _insertar(self.raiz, info)
ABB.insertar = insertar2

arbol = ABB()
for el in arr:
  arbol.insertar(el)
arbol.dibujar()
arbol.raiz.der.num
```

```python
def getKth2(self, k):
  def _getKth(nodo, k):
    if nodo == None:
      return False
    else:
      num_nodos = 1 #+1 por el nodo en el que estoy
      if nodo.izq != None:
        num_nodos += nodo.izq.num

      if num_nodos == k: #el resultado es donde estoy
        self.k = nodo.info
        return True
      elif num_nodos < k:
        return _getKth(nodo.der, k-num_nodos)
      else: #num_nodos > k
        return _getKth(nodo.izq, k)
  #end _getKth
  if _getKth(self.raiz, k):
    return self.k
  return -1

ABB.getKth = getKth2

for i in range(1,11):
  print(arbol.getKth(i))

#Cuanto toma este algoritmo asintoticamente?
#Si el arbol está relativamente balanceado toma O(log(n))
#Si el arbol está sesgado hacia la derecha toma O(n)
arr2 = [i for i in range(1,11)]

arbol2 = ABB()

for el in arr2:
  arbol2.insertar(el)

arbol2.dibujar()
for i in range(1,11):
  print(arbol2.getKth(i))
```

```python
def getKthElement(arr,k):
  arbol = ABB()
  for el in arr:
    arbol.insertar(el)
  arbol.dibujar()
  return arbol.getKth(k)

getKthElement(arr, 10)
```

```python
def getAllKth(self, k):
  k_element = self.getKth(k) #O(log(n))
  if k_element == -1:
    k_element = 2**32 -1 #infinito
  def _getAllKth(nodo, k_element): #O(k)
    if nodo == None:
      return []
    else:
      izq = _getAllKth(nodo.izq, k_element) # Retorna el arreglo con los elementos menores a k
      if nodo.info <= k_element:
        der = _getAllKth(nodo.der, k_element) # Si somos menores al k-esimo, tenemos elementos menores a k
                                              # en nuestro lado derecho.
        res = [*izq, nodo.info, *der]  # Como izq + [nodo.info] + der
        return res
      return izq

  return _getAllKth(self.raiz, k_element)

ABB.getAllKth = getAllKth
# El orden es O(log(n)+k), si k = n => O(log(n) + n) => O(n)
# Si es el peor caso para getKth O(n+n) => O(n)
arbol.getAllKth(15)
```

# Problema 2


En esta pregunta usted debe implementar los métodos "transpose" y "move to front" sobre una lista secuencial implementada sobre un arreglo.





+ Implemente un método `transpose(x,a,n)`,que busque x dentro del arreglo a,que contiene $n$ elementos (ubicados en $a[0],...,a[n-1]$) y al encontrarlo, le aplique una transposición. Si no se encuentra, se debe agregar x al final, y luego se le debe aplicar la transposición.

```python
def transpose(x,a,n):
  # busqueda lineal
  pos = 0
  while pos < n:
    if a[pos] == x:
      break
    pos += 1

  # agregar elemento si no se encontro
  if pos == n:
    a.append(x)

  # transposición
  if pos > 0:
    a[pos], a[pos-1] = a[pos-1], a[pos]
```

```python
# Caso x en a
a1 = [2,6,8,7,4,5,9]
transpose(7, a1, len(a1))
print(a1)

# Caso x no esta en a
transpose(10, a1, len(a1))
print(a1)
```

+ Repita lo anterior, para el método de “move to front”.

```python
def moveToFront(x,a,n):
 # busqueda lineal
  pos = 0
  while pos < n:
    if a[pos] == x:
      break
    pos += 1

  # agregar elemento si no se encontro
  if pos == n:
    a.append(x)

  # mover al primer índice
  while pos > 0:
    a[pos], a[pos-1] = a[pos-1], a[pos]
    pos -= 1
```

```python
# Caso x en a
a1 = [2,6,8,7,4,5,9]
moveToFront(7, a1, len(a1))
print(a1)

# Caso x no esta en a
moveToFront(10, a1, len(a1))
print(a1)
```

## Move to Root en ABB
El algoritmo _autoorganizante_ funciona de la siguiente manera: cuando un nodo es accedido, se corta el árbol por la vertical, obteniendo el nodo y dos subárboles (ej. @peor, luego @corte). Uno cuya raíz es la raíz del árbol original y otro cuya raíz es el hijo izquierdo si el nodo es un hijo izquierdo, o derecho, si es un hijo derecho. Luego, se mueve el nodo a la raíz, resultando el árbol en la @final.



```python
class Nodoe:
    def __init__(self):
        pass  # Empty/null node

    def imprimirPreorden(self, s):
        print(s, '[]')

    def buscar(self, n, padre=None):
        return None, False  # Not found


class Nodoi:
    def __init__(self, izq, info, der):
        self.izq = izq
        self.info = info
        self.der = der

    def imprimirPreorden(self, s):
        print(s, self.info)
        self.izq.imprimirPreorden(s + '  ')
        self.der.imprimirPreorden(s + '  ')

    def buscar(self, n, padre=None):
        if self.info == n:
            return self, True
        elif n < self.info:
            return self.izq.buscar(n, self)
        else:
            return self.der.buscar(n, self)


class Arbol:
    def __init__(self, raiz=None):
        self.raiz = raiz

    # ##########################################################
    # Aquí lo importante de nuestra solución
    # Función útil para imprimir
    def imprimirPreorden(self):
        if self.raiz:
            self.raiz.imprimirPreorden('')
        else:
            print('Árbol vacío')

    def buscar(self, n):
        if self.raiz:
            nodo, encontrado = self.raiz.buscar(n)
            if encontrado:
                self.raiz = self._move_to_front(self.raiz, n)
            return nodo, encontrado
        else:
            return None, False

    def rotate_left(self, parent):
        """El hijo derecho de parent sube (reemplaza a parent)"""
        pivot = parent.der
        parent.der = pivot.izq
        pivot.izq = parent
        return pivot

    def rotate_right(self, parent):
        """El hijo izquierdo de parent sube (reemplaza a parent)"""
        pivot = parent.izq
        parent.izq = pivot.der
        pivot.der = parent
        return pivot

    def _move_to_front(self, node, key):
        if not isinstance(node, Nodoi):
            return node
        if key < node.info:
            if isinstance(node.izq, Nodoi):
                if node.izq.info == key:
                    return self.rotate_right(node)
                node.izq = self._move_to_front(node.izq, key)
                return self.rotate_right(node)
        elif key > node.info:
            if isinstance(node.der, Nodoi):
                if node.der.info == key:
                    return self.rotate_left(node)
                node.der = self._move_to_front(node.der, key)
                return self.rotate_left(node)
        return node  # key is already at root or not found

    def move_to_front(self, n):
        if self.raiz:
            self.raiz = self._move_to_front(self.raiz, n)
    # ##########################################################
```

```python
arbol = Arbol(Nodoi(Nodoi(Nodoi(Nodoi(Nodoe(),
                                    1,
                                    Nodoe()),
                              2,
                              Nodoe()),
                        3,
                        Nodoe()),
                  4,
                  Nodoe()))
```

# Problema 3

Los árboles ABB son una estructura de datos que cumplen las siguientes propiedades:
- Es un árbol binario, es decir, tiene un nodo raíz y a lo más dos subárboles en cada nivel.
- Todos los elementos pertenecientes al subárbol derecho, son mayores que la raíz.
- Todos los elementos pertenecientes al subárbol izquierdo, son menores que la raíz.

Programe una función def `checkABB(nodo)` que, dado un árbol binario, retorne `True` si es de búsqueda o `False` en caso contrario.

```python
def checkABB(nodo):
  # A partir de los padres, cada nodo tiene un rango de valores en el que puede
  # estar su valor
  def recursion(nodo, valor_min = -float('inf'), valor_max = float('inf')):
    if not nodo:
      return True

    if nodo.info < valor_min or nodo.info > valor_max:
      return False

    # El nodo izquierdo y sus hijos deben tener valor menor al nodo actual
    izq_es_abb = recursion(nodo.izq, valor_min, nodo.info)

    # El nodo derecho y sus hijos deben tener valor mayor al nodo actual
    der_es_abb = recursion(nodo.der, nodo.info, valor_max)

    return izq_es_abb and der_es_abb

  return recursion(nodo)
```

```python
# Ejemplos enunciado
# No es abb
abb1 = Nodo(
    5,
    Nodo(2, Nodo(1), Nodo(3)),
    Nodo(6, Nodo(4), Nodo(7)),
)

# Si es abb
abb2 = Nodo(
    4,
    Nodo(2, Nodo(1), Nodo(3)),
    Nodo(6, Nodo(5), Nodo(7)),
)
assert checkABB(abb1) == False and checkABB(abb2) == True
```
